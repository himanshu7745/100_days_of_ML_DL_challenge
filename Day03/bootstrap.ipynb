```python
import numpy as np
import matplotlib.pyplot as plt

# Sample Data
data = np.array([
    10,12,14,16,18,20,22,24
])

B = 1000

bootstrap_means = []
```

```python
# Bootstrap Sampling
for _ in range(B):

    sample = np.random.choice(
        data,
        size=len(data),
        replace=True
    )

    bootstrap_means.append(
        np.mean(sample)
    )

bootstrap_means = np.array(
    bootstrap_means
)

# Statistics
bootstrap_mean = np.mean(
    bootstrap_means
)

bootstrap_std_error = np.std(
    bootstrap_means
)

lower = np.percentile(
    bootstrap_means,
    2.5
)

upper = np.percentile(
    bootstrap_means,
    97.5
)

print("Bootstrap Results")
print("-" * 40)
print("Bootstrap Mean:", bootstrap_mean)
print("Bootstrap Standard Error:", bootstrap_std_error)
print("95% Confidence Interval:")
print((lower, upper))
```

```python
plt.figure(figsize=(8,6))

plt.hist(
    bootstrap_means,
    bins=30
)

plt.axvline(
    lower,
    linestyle="--",
    label="Lower CI"
)

plt.axvline(
    upper,
    linestyle="-",
    label="Upper CI"
)

plt.title("Bootstrap Distribution")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.legend()

plt.show()
```
